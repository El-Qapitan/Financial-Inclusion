```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df = pd.read_csv('Train.csv')
```

```python
df
```

```python
label = LabelEncoder()
label.fit(df.bank_account.drop_duplicates())
df.bank_account = label.transform(df.bank_account)

label.fit(df.location_type.drop_duplicates())
df.location_type = label.transform(df.location_type)

label.fit(df.cellphone_access.drop_duplicates())
df.cellphone_access = label.transform(df.cellphone_access)

label.fit(df.gender_of_respondent.drop_duplicates())
df.gender_of_respondent = label.transform(df.gender_of_respondent)

label.fit(df.marital_status.drop_duplicates())
df.marital_status = label.transform(df.marital_status)

df.head()
```

```python
df.tail()
```

```python
x_lin = df.drop(['bank_account', 'country', 'uniqueid', 'relationship_with_head', 'education_level', 'job_type'], axis = 1)
y_lin = df[['bank_account']]

x_lin_train, x_lin_test, y_lin_train, y_lin_test = train_test_split(x_lin, y_lin, test_size = 0.3, random_state = 40)

Linear_model = LinearRegression()
Linear_model.fit(x_lin_train, y_lin_train)
```

```python
for idx, col_name in enumerate(x_lin_train.columns):
    print('coefficient for {} is {}'.format(col_name, Linear_model.coef_[0][idx]))
```

```python
intercept = Linear_model.intercept_[0]
intercept
```

```python
y_pred = Linear_model.predict(x_lin_test)
pred = pd.DataFrame(x_lin_test)
pred['bank_account'] = y_lin_test
pred['predbank_account'] = y_pred
pred
```

```python
Linear_model.score(x_lin_test, y_lin_test)
```
